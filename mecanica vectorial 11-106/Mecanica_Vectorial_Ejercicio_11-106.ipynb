# Solución con apoyo visual de Python del ejercicio 11.106 de M.V.

Este es un ejercicio de caida libre, que por si no lo saben, es un caso en donde la aceleración en todas las direcciones es constante, por lo tanto sus ecuaciones cinematicas corresponde a las que espuse en la segunda publicación de mi Instagram:

<img src="https://github.com/TheMechanic16/Ejercios-de-Dinamica/blob/master/mecanica%20vectorial%2011-106/cinematica%20acel%20cosnt.jpg?raw=1">

Aquí al fin puede verse porque siempre puse las flechas en las ecuaciones, esto indica que son expresiones vectoriales y es valida en todas las direcciones. Para ilustrar esto escribire las ecuaciones de la posición descompuesta en ejes, que a fin de cuentas la posición es lo que me permitira animar la trayectoria:

\begin{equation}
 \label{eq:posición x}\tag{1}
 s_{x}=x(t)=x_{0}+v_{0_{x}}*(t-t_{0})+\frac{a_{x}}{2}*(t-t_{0})^2
\end{equation}

\begin{equation}
 \label{eq:posición y}\tag{2}
 s_{y}=y(t)=y_{0}+v_{0_{y}}*(t-t_{0})+\frac{a_{y}}{2}*(t-t_{0})^2
\end{equation}

Entendiendo que estas son expresiones generales podemos empezar el codigo importanto las libreria necesarias y definiendo la función general para las posiciones:

```python
#pip install celluloid # ejecutar comando si se usa en Colab.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from celluloid import Camera
from math import pi
from IPython.display import HTML

def pos(t,v0,s0,a):
    p=s0+v0*(t-t0)+(a/2)*(t-t0)**2
    return p
```

Para la solución es totalmente necesario escoger un origen y un sistema de referencía, porque de este dependera la magnitud y dirección de las condiciones iniciales,en mi caso, lo escogere en el suelo desde donde lanza la jugadora y las direcciones positivas tipicas, osea creciente hacia arriba y a la derecha:

<img src="https://github.com/TheMechanic16/Ejercios-de-Dinamica/blob/master/mecanica%20vectorial%2011-106/sistema%20de%20referencia%20y%20origen%20escogido.png?raw=1">

Las condiciones iniciales quedan definidas para este sistema de referencia, aclarando que el cuerpo al que pienso seguir su posición es, logicamente, la pelota:

```python
t0= 0 # [s] tiempo inicial.
distXTablero=-16 # [ft] distancia x inicial del balon al tablero.
s0x= 0 # [ft] posición inicial x del balon.
s0y= 6.8 # [ft] posición inicial y del balon.
din= 9 # [in] distancia en x del tablero al aro en pulgadas.
def dis_Foot(din):
    d=din*0.0833333 # [ft] distancia en x del tablero al aro.
    return d
def x_final(d):
    xf= distXTablero+d # [ft] posición x de la canasta respecto al origen.
    return xf
yf= 10 # [ft] posición y de la canasta respecto al origen.
ax= 0 # [ft/s2] valor aceleración constante en x.
ay= -32.16 # [ft/s2] valor aceleración constante en y.
angOmega= 150*pi/180 # [°] angulo de la velocidad respecto al eje x en el primer cuadrante.
```

La solución del problema se consigue solucionando el sistema de ecuaciones, yo lo hice manualmente y sin tener en cuenta los terminos que se anulaban:

\begin{equation}
 \label{eq:solución}\tag{3}
 v_{0}=\sqrt{\frac{a_{y}*x^2}{2*(y-y_{0}-\tan(\theta)*x)*\cos^2(\theta)}}
\end{equation}

\begin{equation}
 \label{eq:tiempo final}\tag{4}
 t=\frac{x}{v_{0}*\cos(\theta)}
\end{equation}

```python
def velIni(xf):
    v0= np.sqrt((ay*xf**2)/(2*((np.cos(angOmega))**2)*(yf-s0y-np.tan(angOmega)*xf))) # [ft/s] magnitud de la velocidad necesaria para meter el tiro.
    return v0
def time_fin(xf,v0):
    tf=xf/(v0*np.cos(angOmega)) # tiempo que tarda el balon en entrar al aro.
    return tf
```

Esto resuelve el problema del ejercicio, pero como queremos apoyar el ejercicio,creamos las animaciones:

```python
fig, axes = plt.subplots()
xdata, ydata = [], []
axes.set_xlim(-18,2)
axes.set_ylim(0, 15)
camera = Camera(fig)
d=dis_Foot(din)
xf=x_final(d)
v0=velIni(xf)
tf=time_fin(xf,v0)
time = np.linspace(0, tf+0.01, 100)
for t in time:
    x=pos(t,v0*np.cos(angOmega),s0x,ax)
    y=pos(t,v0*np.sin(angOmega),s0y,ay)
    xdata.append(x)
    ydata.append(y)
    tm = axes.text(-2, 13, 'time = %.1fs' % t)
    axes.plot(xdata,ydata,color='blue')
    axes.plot(x,y,'o',color='orange')
    canasta=axes.plot([xf-.5,xf+.5],[yf,yf],'r-')
    tablero=axes.plot([distXTablero,distXTablero],[9,11],'r-')
    conexion1=axes.plot([distXTablero,distXTablero+d],[9.3,yf],'r-')
    conexion2=axes.plot([distXTablero,distXTablero+d],[9.6,yf],'r-')
    camera.snap()
plt.close()
animation = camera.animate(interval=20)
#animation.save('tiro_perfecto.gif',writer='ffmpeg',fps=20) # ejecutar esta linea cuando se quiera generar el recurso.
#animation.save('tiro_perfecto.mp4',writer='ffmpeg',fps=60) # ejecutar esta linea cuando se quiera generar el recurso.
```

Para graficar todas la variables cinematicas tenemos que definir tambien la función de la velocidad respecto al tiempo:

\begin{equation}
 \label{eq:ecuacion de velocidad constante}\tag{5}
 v=v(t)=v_{0}+a*(t-t_{0})
\end{equation}

```python
def velocidad(v0,a,t):
   v=v0+a*t
   return v
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.set_figheight(12)
fig.set_figwidth(14)
xdata, ydata,velY,velX = [], [], [], []
ax1.set_xlim(-18,2)
ax1.set_ylim(0, 15)
ax1.set_title('Tiro')
ax1.set_xlabel('dis_x (ft)')
ax1.set_ylabel('dis_y (ft)')
ax2.set_title('Posición')
ax2.set_xlabel('Tiempo (s)')
ax2.set_ylabel('dis_y (ft)')
ax3.set_title('Velocidad')
ax3.set_xlabel('Tiempo (s)')
ax3.set_ylabel('Velocidad (ft/s)')
ax4.set_title('Aceleración')
ax4.set_xlabel('Tiempo (s)')
ax4.set_ylabel('Aceleración (ft/s^2)')
camera = Camera(fig)
d=dis_Foot(din)
xf=x_final(d)
v0=velIni(xf)
tf=time_fin(xf,v0)
time = np.linspace(0, tf+0.01, 100)
tiempo=[]
aY=[]
aX=[]
ax2.plot(tiempo,ydata,color='red',label='y')
ax2.plot(tiempo,xdata,color='green',label='x')
ax3.plot(tiempo,velY,color='red',label='y')
ax3.plot(tiempo,velX,color='green',label='x')
ax4.plot(tiempo,aY,color='red',label='y')
ax4.plot(tiempo,aX,color='green',label='x')
ax2.legend(loc='upper right')
ax3.legend(loc='upper right')
ax4.legend(loc='upper right')
for t in time:
    x=pos(t,v0*np.cos(angOmega),s0x,ax)
    y=pos(t,v0*np.sin(angOmega),s0y,ay)
    xdata.append(x)
    ydata.append(y)
    ax1.plot(xdata,ydata,color='blue')
    ax1.plot(x,y,'o',color='orange')
    canasta=ax1.plot([xf-.5,xf+.5],[yf,yf],'r-')
    tablero=ax1.plot([distXTablero,distXTablero],[9,11],'r-')
    conexion1=ax1.plot([distXTablero,distXTablero+d],[9.3,yf],'r-')
    conexion2=ax1.plot([distXTablero,distXTablero+d],[9.6,yf],'r-')
    tiempo.append(t)
    ax2.plot(tiempo,ydata,color='red')
    ax2.plot(tiempo,xdata,color='green')
    vy=v0*np.sin(angOmega)
    vx=v0*np.cos(angOmega)
    velY.append(velocidad(vy,ay,t))
    velX.append(velocidad(vx,ax,t))    
    ax3.plot(tiempo,velY,color='red')
    ax3.plot(tiempo,velX,color='green')
    aY.append(ay)
    aX.append(ax)
    ax4.plot(tiempo,aY,color='red')
    ax4.plot(tiempo,aX,color='green')
    camera.snap()
plt.close()
animation11= camera.animate(interval=20)
#animation11.save('variables.gif',writer='pillow',fps=60) # ejecutar esta linea cuando se quiera generar el recurso.
#animation11.save('variables.mp4',writer='ffmpeg',fps=60) # ejecutar esta linea cuando se quiera generar el recurso.
```
